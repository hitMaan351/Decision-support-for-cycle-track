```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as MSE
from statsmodels.tsa.arima.model import ARIMA
import warnings
import random as rd
import math
```

```python
accident_by_month = pd.read_csv(".\\..\dataset\\accident_by_month.csv", sep = ",")
accident_by_month["n_accidents"] = accident_by_month["date.1"]
accident_by_month["n_accidents"] = accident_by_month["n_accidents"].apply(lambda x: int(x))
accident_by_month = accident_by_month.drop(["date.1", "voie.1"], axis = 1)
accident_by_month = accident_by_month. dropna()

```

```python
accident_by_month.info()
```

```python
totaux_accidents = accident_by_month[["voie", "n_accidents"]].groupby(["voie"]).sum()
totaux_accidents = totaux_accidents.sort_values(by="n_accidents", ascending = False)

print(totaux_accidents)
```

```python
streets = list(set(accident_by_month["voie"]))

dates = list(set(accident_by_month["date"]))

print("nombre de rues : ", len(streets))
print("nombre de mois : ", len(dates))

array_no_accidents = []

for idx, month in enumerate(dates):
    for street in streets:
        df_street_and_date = accident_by_month[np.logical_and(accident_by_month["date"] == month, accident_by_month["voie"] == street)]
        if len(df_street_and_date.index) == 0 : #si pas d'accidents, on met 0
            row_no_accidents = [month, street, 0]
            array_no_accidents.append(row_no_accidents)
    print((idx+1) / len(dates)*100, '%')

array_no_accidents = np.array(array_no_accidents)
df_no_accidents= pd.DataFrame(data = array_no_accidents, columns = ["date", "voie", "n_accidents"])


accident_by_month = pd.concat((accident_by_month, df_no_accidents), axis = 0)
accident_by_month.index = range(len(accident_by_month.index))

```

```python
for street in streets:
    if len(accident_by_month[accident_by_month["voie"] == street]) != 36 :
        print(street)

# chaque nom de rue a 36 lignes dans le dataframe
```
